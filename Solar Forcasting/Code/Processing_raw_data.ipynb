```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import os

# Define the directory where your CSV files are located
directory = "D:\Projects\Solar Forcasting\Data"

# Initialize an empty list to store DataFrames
dfs = []

# Loop through all CSV files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        filepath = os.path.join(directory, filename)
        df = pd.read_csv(filepath)
        dfs.append(df)

# Concatenate the DataFrames vertically to create a single DataFrame
merged_df = pd.concat(dfs, ignore_index=True)

# Now, merged_df contains all rows from all CSV files
```

```python
merged_df
```

```python
sorted_df = merged_df.sort_values(by=["Year", "Month", "Day", "Hour", "Minute"])
```

```python
sorted_df = merged_df.copy()
```

```python
# Combine Year, Month, Day, Hour, and Minute columns into a single datetime column
sorted_df["datetime"] = pd.to_datetime(
    sorted_df[["Year", "Month", "Day", "Hour", "Minute"]]
)
```

```python
sorted_df
```

```python
sorted_df.set_index("datetime", inplace=True)
```

```python
sorted_df
```

```python
sorted_df
```

```python
# Define a dictionary for aggregation functions for numerical columns
agg_dict = {
    "Year": "last",
    "Month": "last",
    "Day": "last",
    "Hour": "last",
    "Minute": "mean",
    "Temperature": "mean",
    "DHI": "mean",
    "DNI": "mean",
    "GHI": "mean",
    "Relative Humidity": "mean",
    "Wind Direction": "mean",
    "Wind Speed": "mean",
    "Pressure": "mean",
    "Dew Point": "mean",
}

# Resample the DataFrame to 1-hour frequency and aggregate data
hourly_df = sorted_df.resample("1H").agg(agg_dict)
```
